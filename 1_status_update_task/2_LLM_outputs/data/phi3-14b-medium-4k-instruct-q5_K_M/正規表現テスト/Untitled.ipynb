```python
# created by gpt, https://chatgpt.com/share/b12baa02-5e67-4dd9-aac1-c2f68cabd47d
# 
import pandas as pd
import os
import glob

# 1から20までの数字でnumbersリストを初期化
numbers = [str(i) for i in range(1, 21)]

# カレントディレクトリのすべてのCSVファイルを取得
csv_files = glob.glob("*.csv")

# 処理結果を保持するリスト
dataframes = []

for file in csv_files:
    # CSVファイルを読み込む
    df = pd.read_csv(file)
    
    # result列に指定の処理を適用
    df['result'] = df['result'].apply(lambda x: x if all(number in x for number in numbers) else pd.NA)
    
    # 処理後のデータフレームをリストに追加
    dataframes.append(df)

# # データフレームリストの確認
# for i, df in enumerate(dataframes):
#     print(f"DataFrame {i + 1} from file: {csv_files[i]}")
#     print(df)
#     print("\n")

# マスターフレーム（欠損の無いdfを作成する）
master_df = dataframes[1]
for df in dataframes:
    # ちなみに下記のcombineはmaster_dfがNaNの時だけdfの値がmaster_dfに入る。
    master_df['result'] = master_df['result'].combine_first(df['result'])
    
# マスターフレームに欠損がないかを確認する。なければ、補完が成功している。
filtered_df = master_df[~master_df['result'].apply(lambda x: all(number in x for number in numbers))]
print(filtered_df)

# 最後に、各データフレームの欠損値をmaster_dfで補完していく。
for df in dataframes:
    df['result'] = df['result'].combine_first(master_df['result'])
    
    
# 次に、テキストたちを連結していく
def concatenate_text(text):
    # ナンバリングとピリオドを削除
    text = re.sub(r'\d+\.\s*', '', text)
    # ダブルクオーテーションを削除
    text = re.sub(r'"', '', text)
    # 改行を削除
    text = text.replace('\n', ' ')
    return text

for df in dataframes:
    # 'result' 列に関数を適用
    df['result'] = df['result'].apply(concatenate_text)

# 最後の最後にdataframesたちをカレントディレクトリ直下にフォルダを作成し、保存する。
# 保存用ディレクトリの作成
output_dir = "processed_csv_files"
os.makedirs(output_dir, exist_ok=True)

# 各データフレームを保存
model_name = input("model_name?> ")
for i, df in enumerate(dataframes):
    output_path = os.path.join(output_dir, f"processed_result{i+1}_{model_name}.csv")
    df.to_csv(output_path, index=False)

print(f"Processed files have been saved to the directory: {output_dir}")
```
